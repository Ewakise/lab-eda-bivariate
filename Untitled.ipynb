```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency, pearsonr, probplot

# Load Data
df = pd.read_csv("amz_uk_price_prediction_dataset.csv")

#Quick Overview
df.info()
df.head()
```

```python
# Crosstab Analysis
crosstab = pd.crosstab(df['category'], df['isBestSeller'], normalize='index') * 100
crosstab_sorted = crosstab.sort_values(True, ascending=False)
crosstab_sorted.head()
```

```python
# Chi-square Test
contingency = pd.crosstab(df['category'], df['isBestSeller'])
chi2, p, dof, expected = chi2_contingency(contingency)

# Cram√©r's V
n = contingency.sum().sum()
phi2 = chi2 / n
r, k = contingency.shape
cramers_v = np.sqrt(phi2 / min(k - 1, r - 1))

chi2, p, cramers_v
```

```python
# Stacked Bar Chart
crosstab_plot = pd.crosstab(df['category'], df['isBestSeller'])
crosstab_plot.div(crosstab_plot.sum(1), axis=0).plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title("Best-Seller Proportion by Category")
plt.ylabel("Proportion")
plt.xlabel("Category")
plt.legend(title="Is Best Seller")
plt.tight_layout()
plt.show()
```

```python
#Remove Price Outliers (IQR Method)
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_no_outliers = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
```

```python
# üéª Violin Plot of Price Distribution (Top 20 Categories)
top_categories = df_no_outliers['category'].value_counts().nlargest(20).index

plt.figure(figsize=(15, 6))
sns.violinplot(data=df_no_outliers[df_no_outliers['category'].isin(top_categories)], x='category', y='price')
plt.xticks(rotation=90)
plt.title("Price Distribution Across Top 20 Categories")
plt.tight_layout()
plt.show()
```

```python
# Category with Highest Median Price
median_prices = df_no_outliers.groupby('category')['price'].median().sort_values(ascending=False)
median_prices.head(1)
```

```python
# üìä Bar Chart: Average Price of Top 10 Categories
top_10_categories = df_no_outliers['category'].value_counts().nlargest(10).index
avg_prices = df_no_outliers[df_no_outliers['category'].isin(top_10_categories)].groupby('category')['price'].mean()

avg_prices.sort_values(ascending=False).plot(kind='bar', figsize=(10, 6))
plt.title("Average Price of Top 10 Categories")
plt.ylabel("Average Price")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# üì¶ Box Plot: Ratings by Category (Top 10)
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_no_outliers[df_no_outliers['category'].isin(top_10_categories)], x='category', y='stars')
plt.title("Product Ratings by Category (Top 10)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Category with Highest Median Rating
median_ratings = df_no_outliers.groupby('category')['stars'].median().sort_values(ascending=False)
median_ratings.head(1)
```

```python
#Correlation Coefficient Between Price and Rating
correlation, p_value = pearsonr(df_no_outliers['price'], df_no_outliers['stars'])
correlation, p_value
```

```python
#Scatter Plot: Price vs Rating
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_no_outliers, x='stars', y='price', alpha=0.3)
plt.title("Price vs Rating")
plt.tight_layout()
plt.show()
```

```python
#Heatmap of All Numeric Correlations
numerical_cols = ['price', 'stars', 'reviews', 'boughtInLastMonth']
corr_matrix = df_no_outliers[numerical_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
#QQ Plot: Price Normality Check
plt.figure(figsize=(6, 6))
probplot(df_no_outliers['price'], dist="norm", plot=plt)
plt.title("QQ Plot of Product Prices")
plt.tight_layout()
plt.show()
```

```python
summary = {
    "Top Categories by Best-Seller Proportion": crosstab_sorted[True].head().to_dict(),
    "Chi-square Test p-value": p,
    "Cram√©r's V": cramers_v,
    "Category with Highest Median Price": median_prices.idxmax(),
    "Category with Highest Average Price (Top 10)": avg_prices.idxmax(),
    "Category with Highest Median Rating": median_ratings.idxmax(),
    "Price-Rating Correlation Coefficient": correlation,
    "Price-Rating Correlation p-value": p_value
}

for key, value in summary.items():
    print(f"{key}:")
    print(value, "\n")
```
